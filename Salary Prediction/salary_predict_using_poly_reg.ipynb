Import Libraries


```python
import pandas as pd
```

Load Dataset

```python
dataset=pd.read_csv('/Users/deepikakala/Documents/Projects/ML Project/Salary Prediction/dataset.csv')
```

Summarizing the dataset

```python
print(dataset.shape)
print(dataset.head(5))
```

Segregating the dataset into X & Y

```python
X=dataset.iloc[:,:-1].values
X
```

```python
Y=dataset.iloc[:,-1].values
Y
```

Training Dataset Using Linear Regression

```python
from sklearn.linear_model import LinearRegression
modelLR=LinearRegression()
modelLR.fit(X,Y)
```

Visualizing the Linear Regression Results

```python
import matplotlib.pyplot as plt
plt.scatter(X, Y, color="blue")
plt.plot(X, modelLR.predict(X))
plt.title("Linear Regression")
plt.xlabel('Level')
plt.ylabel('Price')
```

fit()-Training model-Calculating initial 
transfor()-After Training we gonna transform Data by using above calculated values
fit_transform()-First fit & transform


Covert X to Polynomial format(X^n)

n-degree

n=2 consists x & x^2

```python
from sklearn.preprocessing import PolynomialFeatures
modelPR=PolynomialFeatures(degree=5)
xPoly=modelPR.fit_transform(X)
```

Train same linear reg with X-Polynimial instead X

```python
modelPLR=LinearRegression()
modelPLR.fit(xPoly, Y)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, Y, color="blue")
plt.plot(X, modelPLR.predict(modelPR.fit_transform(X)))
plt.title("Polynomial Regression")
plt.xlabel('Level')
plt.ylabel('Price')
```

Prediction Using Polynomial Regression

```python
X=4
salaryPred=modelPLR.predict(modelPR.fit_transform([[X]]))
print('Salary of a person with level {0} is {1}'.format(X, salaryPred))
```
